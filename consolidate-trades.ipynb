```python

import IPython
import pandas as pd
import numpy as np

from datetime import datetime, date
from math import *
import json
%matplotlib inline
from decimal import *
getcontext().prec = 6
import re
from decimal import Decimal

from os import listdir
from os.path import isfile, join
```

```python
def read_binance_transactions(year):
    filename = f'./trading-data/binance/trading-history-{year}.csv'
    re_pair = re.compile('(?P<base>[a-z]+)(?P<quote>GBP|USDT|ETH|BTC|USDC|DAI|BUSD)', re.IGNORECASE)
    re_curr_amout = re.compile('^(?P<amt>[\d\.,]+)(?P<curr>\w+)$')

    df = pd.read_csv(filename, index_col=0, parse_dates=True)
    df['Exchange'] = 'Binance'
    df.rename(columns={'Date(UTC)': 'DateUTC'}, inplace=True)
    df[['BaseCurr', 'QuoteCurr']] = df.Pair.str.extract(re_pair, expand=True)
    df[['ExecutedAmt', 'ExecutedCurr']]  = df.Executed.str.split(re_curr_amout, expand=True).iloc[:,[1,2]]
    df[['AmountAmt', 'AmountCurr']]  = df.Amount.str.split(re_curr_amout, expand=True).iloc[:,[1,2]]
    df[['FeeAmt', 'FeeCurr']]  = df.Fee.str.split(re_curr_amout, expand=True).iloc[:,[1,2]]
    # df.index = df.index.map(lambda i: f'binance_{year}_{i}')
    return df
# df.csv
```

```python
df_binance_all_trades = pd.concat([
    # read_binance_transactions(2020),
    # read_binance_transactions(2021),
    read_binance_transactions(2022)
])
df_binance_all_trades.to_csv('./trading-data/binance-all-trades.csv')
```
